```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk
import random
import time
import threading
import pygame

audio_random = "random.wav"
audio_win = "win.wav"
confetti_img = "confetti1.jpg"

running = False  # Flag untuk mengontrol proses
random_sound = None  # Variabel untuk menyimpan objek suara random.wav

def play_wav(file_path, loop=False):
    """Memutar file WAV. Jika loop=True, suara akan diputar terus-menerus."""
    global random_sound
    pygame.mixer.init()
    sound = pygame.mixer.Sound(file_path)
    if loop:
        sound.play(loops=-1)  # Putar dalam mode loop
        random_sound = sound
    else:
        sound.play()

def stop_wav():
    """Menghentikan suara yang sedang diputar."""
    global random_sound
    if random_sound:
        random_sound.stop()
        random_sound = None

def slot_machine(range_min, range_max):
    global running
    while running:
        # Animasi angka acak (0-9 untuk setiap digit)
        slots = [random.randint(0, 9) for _ in range(4)]
        label_result.config(text=" | ".join(map(str, slots)), font=("Helvetica", 70))
        root.update()
        time.sleep(0.02)

    # Hasil akhir: angka dalam range range_min hingga range_max
    # Konversi range_min dan range_max ke integer untuk validasi
    min_value = int(str(range_min).zfill(4))
    max_value = int(str(range_max).zfill(4))

    # Hasilkan angka akhir dalam range range_min hingga range_max
    while True:
        slots = [random.randint(0, 9) for _ in range(4)]
        final_number = int("".join(map(str, slots)))
        if min_value <= final_number <= max_value:
            break

    # Tampilkan hasil akhir
    label_result.config(text=" | ".join(map(str, slots)))
    result = f"Selamat utk nomor:\n{' | '.join(map(str, slots))}"
    custom_messagebox("Winner", result, 50)

def toggle_slot():
    global running
    if not running:
        running = True
        btn_start.config(text="Stop")
        threading.Thread(target=start_slot).start()
    else:
        running = False
        btn_start.config(text="Mulai")
        stop_wav()  # Hentikan suara random.wav
        play_wav(audio_win)  # Mainkan suara kemenangan

def start_slot():
    play_wav(audio_random, loop=True)  # Putar suara random.wav dalam mode loop
    slot_machine(1, 3630)

def custom_messagebox(title, message, font_size=12):
    msg_window = tk.Toplevel()
    msg_window.title(title)

    img = Image.open(confetti_img)
    photo = ImageTk.PhotoImage(img)

    label_gambar = tk.Label(msg_window, image=photo)
    label_gambar.image = photo
    label_gambar.pack(expand=1)

    label = tk.Label(msg_window, text=message, font=("Helvetica", font_size))
    label.pack(padx=20, pady=20)

    ok_button = tk.Button(msg_window, text="OK", command=msg_window.destroy, font=("Helvetica", 30))
    ok_button.pack(pady=10)

def center_window(root, width=500, height=400):
    screen_width = root.winfo_screenwidth()
    screen_height = root.winfo_screenheight()
    x = (screen_width // 2) - (width // 2)
    y = (screen_height // 2) - (height // 2)
    root.geometry(f"{width}x{height}+{x}+{y}")

# Membuat GUI dengan Tkinter
root = tk.Tk()
root.title("Undian Kupon Berhadiah Paskah 2025 Sek Pniel")
root.geometry("500x400")

frame = tk.Frame(root)
frame.pack(pady=20)

label_result = tk.Label(frame, text="Undian Kupon Berhadiah \nPerayaan Paskah 2025 \nSektor Pniel", font=("Helvetica", 30))
label_result.pack()

btn_start = tk.Button(frame, text="Mulai", font=("Helvetica", 50), command=toggle_slot)
btn_start.pack(pady=20)

center_window(root, 500, 400)
root.mainloop()
```
